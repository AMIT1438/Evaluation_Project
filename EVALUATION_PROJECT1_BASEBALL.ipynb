- Checking Variance Influence Factor

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(x):
    vif=pd.DataFrame()
    vif["VIF FACTOR"]=[variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
    vif['Features']=x.columns
    print(vif)
```

```python
calc_vif(x)
```

- ERA, ER and RA are highly correlated to each other they have highest vif factor as well
- Since ERA is contributing (-83%) which is higer than ER (-82%) and RA (-82%) hence we will drop ER and RA features
- AB, R and 2B have higher vif factor out of which AB is contribution to W only -0.086 which is negligible 
- R is contributing to W at 39% and 2B is contribution to W 38% so we will drop AB and 2B features as well

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LinearRegression
lr=LinearRegression()

from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error #For checking loss function (errors)

from sklearn.model_selection import train_test_split
```

##### Checking the best random state for splitting the the data in train and test

```python
rstate=0
rscore=0

for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=.2,random_state=i)
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    if(round(r2_score(y_test,pred)*100)>rscore):
        rstate=i
        rscore=round(r2_score(y_test,pred)*100)

print('best random state is',rstate," ",rscore)
```

- Accuracy Score is 97% at random state 225 in LinearRegression which is good so we will use 225 as randomstate

```python
#assigning values of x1 (skewremoved values) to x
x=x1
#splitting the data in train and test variables
x_train,x_test,y_train,y_test=train_test_split(x1,y,test_size=.2,random_state=225)
```
