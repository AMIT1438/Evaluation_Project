#### Since there are so many null values so we need to either treat these or drop the columns with maximum number of null values 

```python
#Filling null values with mean in numeric columns and with mode in string (object) columns
df_old=df.copy() #Keeping a copy of original dataframe that may be required later 

for i in df.columns:
    if(df.dtypes[i]=='object'):
        df[i].fillna(df[i].mode(),inplace=True)
    else:
        df[i].fillna(df[i].mean(),inplace=True)
    
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#since Some " " , NaN and np.nan values are still not filled so we will fill these

#Filling Mode values in place of " ", NaN and np.nan

for i in df.columns:
    #if(df.dtypes[i]=='O'):
    mode = df[df[i]!=" "][i].mode()[0]
    df[i] = df[i].replace(" ",mode) 
    df[i] = df[i].replace("NaN",mode)
    df[i] = df[i].replace(np.nan,mode)
  
```
